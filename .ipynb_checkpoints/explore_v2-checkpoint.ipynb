```python
Om_ref = 0.32
Ob_ref = 0.049
wDE_ref = -1.0
h_ref = 0.67
ns_ref = 0.96
As_ref = 2.1e-09
#Onu_ref = 0.058/(93.15*h_ref**2)
z_max_pk = 200       # highest redshift involved

k_per_decade = 35#180     # number of k values, controls final resolution
P_k_max_h_Mpc = 1.0   # this value is directly the maximum k value in the figure in Mpc
```

```python
cosmo_ref = Class()
cosmo_ref.set(
        {'output':'mPk',
          'Omega_Lambda': 0,
          'w0_fld':'-1.1',
          'wa_fld': '-0.2',
          'A_s':As_ref,
          'n_s':ns_ref,
          #'cs2_spline_regulator': 'tanh',
          'P_k_max_h/Mpc' : P_k_max_h_Mpc,
          'z_pk':'0.0, 1.0, 2.0',
          #'background_verbose': 3,
          #'perturbations_verbose':7,
          'evolver': 0,
          'h': h_ref,
          'Omega_b':Ob_ref,
          'Omega_cdm': Om_ref - Ob_ref})#,0.003587746997724588,#0.0013878391714135642,
cosmo_ref.compute()
```

```python
z = cosmo_ref.get_background()['z']
a = 1./(1.+z)

bg = cosmo_ref.get_background()
```

```python
wlog10a_vals_list = ','.join(map(str, np.concatenate((np.array([-14, -3.5]), np.linspace(-3., 0, 8))))) 
wvals_list = ','.join(map(str, np.concatenate((np.array([-1., -1.]), np.random.uniform(-2, -0.5, 8))))) 

cosmo_var = Class()
cosmo_var.set(
        {'output':'mPk',
          'Omega_Lambda': 0.,
          'Omega_fld': 0.,
          'Omega_wspl': -1,
          'use_ppf': 'yes',
          'A_s':As_ref,
          'n_s':ns_ref,
          #'wlog10a_vals': '-14., -1.5, -1., 0.',
          'wlog10a_vals': wlog10a_vals_list,
          #'w_vals': '-1, -.9, -.8, -1.1',
          'w_vals' : wvals_list,
          #'z_pk':'0.0, 1.0, 2.0',
          'evolver': 0,
          #'background_verbose': 5,
          'z_pk': '0., 0.5',
          'h': h_ref,
          'Omega_b':Ob_ref,
          'Omega_cdm': Om_ref - Ob_ref})#,0.003587746997724588,#0.0013878391714135642,
cosmo_var.compute()
```

# $c_{s}^{2} = A \times w(z) + B$

```python
plt.figure(figsize=(7,5))
plt.semilogx(a, cosmo_ref.get_background()['H [1/Mpc]']/cosmo_var.get_background()['H [1/Mpc]'])

plt.xlim(1e-3, 1)
#plt.ylim(1e-4, 1e2)
plt.grid()
```

```python
plt.figure(figsize=(9,6))

plt.semilogx(a, -1*np.ones(len(a)), '--', color='black', label=r'$w_{\rm DE}=-1$ ($\Lambda$CDM)')
plt.semilogx(a, cosmo_var.get_background()['wspl'], label=r'$w_{\rm DE}^{\rm spline}$')
plt.semilogx(a, cosmo_var.get_background()['dwspl_dlna'], label=r'$\frac{d w_{\rm DE}^{\rm spline}}{d \ln a}$')
plt.semilogx(a, cosmo_var.get_background()['int_wspl_dlna'], label=r'$\int w^{\rm spline}_{\rm DE}~ d~ \ln a$')
plt.xlim(1e-4, 1)
plt.ylim(-2, 3)
plt.xlabel(r'$a$')
plt.legend(ncol=2)

#plt.grid()
```

```python
# get P(k) at redhsift z=0
import numpy as np
kk = np.logspace(-3,np.log10(1), 200) # k in h/Mpc
Pk_0 = [] # P(k) in (Mpc/h)**3
Pk_05 = [] # P(k) in (Mpc/h)**3
h = cosmo_var.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk:
    Pk_0.append(cosmo_var.pk(k*h_ref,0.)*h_ref**3) # function .pk(k,z)
    Pk_05.append(cosmo_var.pk(k*h_ref,0.5)*h_ref**3) # function .pk(k,z)
```

```python
plt.figure(figsize=(7,5))
plt.loglog(kk, np.array(Pk_0))
plt.loglog(kk, np.array(Pk_05))

plt.grid()
```
